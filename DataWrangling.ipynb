```python
import pandas as pd

url = "https://raw.githubusercontent.com/giovanni-lucarelli/basket/main/games.csv"
df = pd.read_csv(url)
```

```python
print(df.head())
```

```python
df.dropna(inplace=True)
```

```python
import pandas as pd
import random

# Pick a random HOME_TEAM_ID
random_team_id = random.choice(df['HOME_TEAM_ID'].unique())
print(f"Selected team ID: {random_team_id}")

# Filter games where the team is either home or visitor
team_games = df[
    (df['HOME_TEAM_ID'] == random_team_id) | (df['VISITOR_TEAM_ID'] == random_team_id)
]

# Order by game date
team_games_sorted = team_games.sort_values(by='GAME_DATE_EST')

# Get the first six games
first_six_games = team_games_sorted.head(6)

# Display the dataframe
first_six_games
```
